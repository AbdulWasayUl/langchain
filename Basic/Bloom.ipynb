```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
import torch
torch.set_default_tensor_type(torch.cuda.FloatTensor)
```

```python
model = AutoModelForCausalLM.from_pretrained("bigscience/bloom-1b7", use_cache=True)
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloom-1b7")
```

```python
model.__class__.__name__
```

```python
blog_post_title = 'Transformer Architecture'
```

```python
prompt = f'This is a blog post on {blog_post_title} \n {blog_post_title}'
```

```python
inputs = tokenizer(prompt, return_tensors="pt").to(0)
```

```python
sample = model.generate(**inputs, max_length=200, num_beams=2, num_beam_groups=2, top_k=1, temperature=0.9, repetition_penalty=3.0, diversity_penalty=0.9)
```

```python
print(tokenizer.decode(sample[0], truncate_before_pattern=[r"\n\n^#", "^'''", "\n\n\n"]))
```
