```python
from langchain import HuggingFaceHub, LLMChain
from langchain.prompts import PromptTemplate
```

```python
from google.colab import userdata
import os
os.environ['HUGGINGFACEHUB_API_TOKEN'] = userdata.get('HUGGINGFACEHUB_API_TOKEN')
```

```python
llm = HuggingFaceHub(repo_id="abdelhalim/Rec_Business_Names",
                      model_kwargs={
                         "temperature":0.8
                     }
                     )
# llm = HuggingFaceHub(repo_id="gpt2",
#                      model_kwargs={
#                          "temperature":0.8
#                      }
#                      )

prompt = PromptTemplate(
    input_variables=["cuisine"],
    template="following is a comma seperated list of 5 {cuisine} restaurant names"
)

chain = LLMChain(prompt=prompt, llm=llm, verbose=True)
```

```python
chain.run("mexican")
```
